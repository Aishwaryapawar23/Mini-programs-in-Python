# Prorgam to build a shopping cart.

```python
#inventory
Biscuits=30
Cereals=50
Maggi=16
Oats=0
Cakes=25

#cost per unit
cost_Biscuits=15.5
cost_Cereals=200
cost_Maggi=12
cost_Oats=30
cost_Cakes=35

def add_to_basket():
    print('--------------------Welcome to SHOPIFY------------------------------------')
    print("----------------Available Items of our store -----------------------------")
    print("%s %10s %11s %13s" %("Sr.No","Items","Quantity","Price/Unit"))
    print("%d %16s %9d %10.2f" %(1,"Biscuits",Biscuits,cost_Biscuits))
    print("%d %15s %10d %10.2f" %(2,"Cereals",Cereals,cost_Cereals)) 
    print("%d %15s %10d %10.2f" %(3,"Maggi",Maggi,cost_Maggi)) 
    print("%d %15s %10d %10.2f" %(4,"Oats",Oats,cost_Oats)) 
    print("%d %15s %10d %10.2f" %(5,"Cakes",Cakes,cost_Cakes)) 
    
    
  
    qty_Biscuits=int(input("Enter Quantity of Biscuits:"))   
    if (Biscuits<qty_Biscuits):
        print(' Sorry, This item is out of stock! ')
        qty_Biscuits=0
    else:
        print('Item added to cart.')
        
    qty_Cereals=int(input("Enter Quantity of Cereals:"))
    if (Cereals<qty_Cereals):
        print(' Sorry, This item is out of stock! ')
        qty_Cereals=0
    else:
        print('Item added to cart.')
   
        
    qty_Maggi=int(input("Enter Quantity of Maggi:"))
    if (Maggi<qty_Maggi):
        print(' Sorry, This item is out of stock! ')
        qty_Maggi=0
    else:
        print('Item added to cart.')
   
    qty_Oats=int(input("Enter Quantity of Oats:"))
    if (Oats<qty_Oats):
        print(' Sorry, This item is out of stock! ')
        qty_Oats=0
    else:
        print('Item added to cart.')
    
        
    qty_Cakes=int(input("Enter Quantity of Cakes:"))
    if (Cakes<qty_Cakes):
        print(' Sorry, This item is out of stock! ')
        qty_Cakes=0
    else:
        print('Item added to cart.')
  
    
    
    return qty_Biscuits,qty_Cereals,qty_Maggi,qty_Oats,qty_Cakes

def cal_total_item_cost(qty_Biscuits,qty_Cereals,qty_Maggi,qty_Oats,qty_Cakes):
    total_cost_items=qty_Biscuits*cost_Biscuits+qty_Cereals*cost_Cereals+qty_Maggi*cost_Maggi+qty_Oats*cost_Oats+qty_Cakes*cost_Cakes
    
    return total_cost_items

#If delivery distance is under 4 km - No delivery cost if over 4 km - Delivery Cost = 100

def get_delivery_details():
    print("Enter your delivery details:")
    name=input("Enter your name:")
    address=input("Enter your address:")
    distance=int(input("Distance from Store:"))
    
    return name,address,distance

def cal_delivery_charge(distance):
    if distance <= 4:
        delivery_charge=0
    else:
        delivery_charge=100
        
    return delivery_charge
    
def cal_final_bill(total_cost_items,delivery_charge):
    final_bill_amount=total_cost_items+delivery_charge
    return final_bill_amount
    
    
def generate_bill(qty_Biscuits,qty_Cereals,qty_Maggi,qty_Oats,qty_Cakes,total_cost_items,final_bill_amount,name,address):
    print("--------------------------Your Final Bill -----------------------------------")
    print("%s %10s %10s %10s %10s" %("Sr.No","Items","Quantity","Price/Unit","Total Cost"))
    print("%d %16s %9d %10.2f %10.2f" %(1,"Biscuits",qty_Biscuits,cost_Biscuits,qty_Biscuits*cost_Biscuits))
    print("%d %15s %10d %10.2f %10.2f" %(2,"Cereals",qty_Cereals,cost_Cereals,qty_Cereals*cost_Cereals)) 
    print("%d %15s %10d %10.2f %10.2f" %(3,"Maggi",qty_Maggi,cost_Maggi,qty_Maggi*cost_Maggi))
    print("%d %15s %10d %10.2f %10.2f" %(4,"Oats",qty_Oats,cost_Oats,qty_Oats*cost_Oats))
    print("%d %15s %10d %10.2f %10.2f" %(5,"Cakes",qty_Cakes,cost_Cakes,qty_Cakes*cost_Cakes))
    
    print("Total Cost:",total_cost_items)
    print("Total Bill :" ,total_cost_items,' + ', delivery_charge,'(Delivery Charges) =',final_bill_amount)
    print("Name:",name)
    print("Address:",address)
    print('Distance:',distance)
    
while True:
    
    qty_Biscuits,qty_Cereals,qty_Maggi,qty_Oats,qty_Cakes=add_to_basket()
    
    total_cost_items=cal_total_item_cost(qty_Biscuits,qty_Cereals,qty_Maggi,qty_Oats,qty_Cakes)
    
    # Inventory Update
    
    Biscuits-=qty_Biscuits
    Cereals-=qty_Cereals
    Maggi-=qty_Maggi
    Oats-=qty_Oats
    Cakes-=qty_Cakes
    
    name,address,distance = get_delivery_details()
    
    delivery_charge=cal_delivery_charge(distance)
    
    final_bill_amount=cal_final_bill(total_cost_items,delivery_charge)
    generate_bill(qty_Biscuits,qty_Cereals,qty_Maggi,qty_Oats,qty_Cakes,total_cost_items,final_bill_amount,name,address)
    
    choice=input("Would you like to continue (yes/no)?")
    if choice =="yes":
        continue
    else:
        print()
        print("Thanks for shopping from SHOPIFY. Hope to see you again!")
        break
    
```
