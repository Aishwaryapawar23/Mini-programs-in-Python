# THE CONVERSION APP

### Create a conversion app using functions with parameters and returning their values. The program should be able t perform all the task listed on page. Minimize the use of variables where possible. The output should appear in proper format.

```python
import math
def getinput():
    a=int(input('Enter the value:'))
    return a

def getstringinput():
    b=input('Enter the value:')
    return b

def twoinput():
    x=int(input('Enter 1st number:'))
    y=int(input('Enter 2nd number:'))
    return x,y

def floatvar():
    z=float(input('Enter the value:'))
    return z
    

def Radiantodegree(a):
    print ('Degree of',a,'is',math.radians(a))
    print('------------------------------------------------------------------------')

def Degreetoradian(a):
    print('Radian of',a,'is',math.degrees(a))
    print('------------------------------------------------------------------------')
    
def Fact(a):
    print('Factorial of',a,'is', math.factorial(a))
    print('------------------------------------------------------------------------')

def squareroot(a):
    print('Square root of',a,'is', math.sqrt(a))
    print('------------------------------------------------------------------------')
    
def megabytetonibble(a):
    print('Nibble conversion of',a,'is',a*2000000)
    print('------------------------------------------------------------------------')

def fractiontodecimal(x,y):
    dec = x/y
    print('The decimal value is:',dec)
    print('------------------------------------------------------------------------')
    
def km_to_m(a):
    print('The m/s conversion is:', a*5/18)
    print('------------------------------------------------------------------------')

def m_to_km(a):
    print('The m/s conversion is:',18*a/5)
    print('------------------------------------------------------------------------')
    
def int_to_float(a):
    print('The float value of',a,'is',float(a))
    print('------------------------------------------------------------------------')

def float_to_int(a):
    print('The integer value of',a,'is',int(a))
    print('------------------------------------------------------------------------')

def decimal_to_octal(b):
    print('The octal value of',b,'is',int(b,8))
    print('------------------------------------------------------------------------')

def octal_to_decimal(a):
    print('The decimal value of',a,'is',oct(a))
    print('------------------------------------------------------------------------')
    
def Decimal_to_hexadecimal(b):
    print('The hexadecimal value of',b,'is',int(b,16))
    print('------------------------------------------------------------------------')
    
def Hexadecimal_to_decimal(a):
    print('The hexadecimal value of',a,'is',hex(a))
    print('------------------------------------------------------------------------')
    
def Decimal_to_binary(a):
    print('The binary value of',a,'is',bin(a).replace('b','0'))
    print('------------------------------------------------------------------------')
    
def Binary_to_decimal(b):
    print('The Decimal value of',b,'is',int(b,2))
    print('------------------------------------------------------------------------')
    
def Character_to_ascii(b):
    print('The ascii value of',b,'is',ord(b))
    print('------------------------------------------------------------------------')
    
def ascii_to_char(a):
    print('The character of',a, 'ascii value is',chr(a))
    print('------------------------------------------------------------------------')
    
def Celsius_to_fahrenheit(a):
    print(a,'Celsius=',((a-32)*5/9),'Fahrenheit')
    print('------------------------------------------------------------------------')
    
def Fahrenheit_to_celsius(a):
    print(a,'Fahrenheit=',(a*1.8 + 32),'Celsius')
    print('------------------------------------------------------------------------')
    
def main():
    while True:
        print()
        print('CONVERSION APPLICATION')
        print('------------------------------------------------------------------------')
        print('------------------------------------------------------------------------')
        
        print('1.Radians to degree                      2.Degree to radians')
        print('3.Number to its factorial                4.Number to its square root')
        print('5.Megabyte to nibble                     6.Fraction to decimal')
        print('7.km/h to m/s                            8.m/s to km/h')
        print('9.Integer to float                       10.Float to integer')
        print('11.Decimal to octal                      12.Octal to decimal')
        print('13.Decimal to hexadecimal                14.Hexadecimal to decimal')
        print('15.Decimal to binary                     16.Binary to decimal')
        print('17.Character to ascii                    18.Ascii to character')
        print('19.Celsius to fahrenheit                 20.Fahrenheit to celsius')
        print('Press 0 to Exit')
        print('------------------------------------------------------------------------')

        choice=int(input('Select the type of conversion you want to do:'))
        if choice==1:
            print('Radians to degree')
            c=getinput()
            Radiantodegree(c)

        elif choice==2:
            print('Degree to radians')
            c=getinput()
            Degreetoradian(c)

        elif choice==3:
            print('Number to its factorial')
            c=getinput()
            Fact(c)

        elif choice==4:
            print('Number to its square root')
            c=getinput()
            squareroot(c)    

        elif choice==5:
            print('Megabyte to nibble')
            c=getinput()
            megabytetonibble(c)

        elif choice==6:
            print('Fraction to decimal')
            j,k=twoinput()
            fractiontodecimal(j,k)

        elif choice==7:
            print('km/h to m/s')
            c=getinput()
            km_to_m(c)

        elif choice==8:
            print('m/s to km/h')
            c=getinput()
            m_to_km(c)

        elif choice==9:
            print('Integer to float')
            c=getinput()
            int_to_float(c)

        elif choice==10:
            print('Float to integer')
            e=floatvar()
            float_to_int(e)

        elif choice==11:
            print('Decimal to octal')
            d=getstringinput()
            decimal_to_octal(d)

        elif choice==12:
            print('Octal to decimal')
            c=getinput()
            octal_to_decimal(c)

        elif choice==13:
            print('Decimal to hexadecimal')
            d=getstringinput()
            Decimal_to_hexadecimal(d)

        elif choice==14:
            print('Hexadecimal to decimal')
            c=getinput()
            Hexadecimal_to_decimal(c)
            
        elif choice==15:
            print('Decimal to binary ')
            c=getinput()
            Decimal_to_binary(c)

        elif choice==16:
            print('Binary to decimal')
            d=getstringinput()
            Binary_to_decimal(d)

        elif choice==17:
            print('Character to ascii')
            d=getstringinput()
            Character_to_ascii(d)

        elif choice==18:
            print('Ascii to character')
            c=getinput()
            ascii_to_char(c)

        elif choice==19:
            print('Celsius to fahrenheit')
            c=getinput()
            Celsius_to_fahrenheit(c)


        elif choice==20:
            print('Fahrenheit to celsius')
            c=getinput()
            Fahrenheit_to_celsius(c)

        elif choice==0:
            print('Thankyou for using conversion app!')
            break
        else:
            print('Enter valid input!')
            


main()
```
