# Create MCQ Quiz with minimum 10 questions

### Accept users response for all questions

### Evaluate students response to multiple choice quiz questions and provides the total marks that the student scored.

### You have been provided an answer key that contains the correct options for all questions.As input for every question the user will provide an option as their correct choice.

```python
q1="""1.Is Python case sensitive?
a) no
b) yes
c) machine dependent
d) none of the mentioned"""

q2="""2.Which of the following is the correct extension of the Python file?
a) .python
b) .pl
c) .py
d) .p"""

q3="""3.Which of the following character is used to give single-line comments in Python?
a) //
b) #
c) !
d) /* """

q4="""4.Which of the following is a Python tuple?
a) {1, 2, 3}
b) {}
c) [1, 2, 3]
d) (1, 2, 3)"""

q5="""5.Which one of these is floor division?
a) /
b) //
c) %
d) None of the mentioned"""

q6="""6. What is a correct syntax to output "Hello World" in Python?.
a) echo('Hello world')
b) print('Hello world')
c) p('Hello world')
d) echo'Hello world"""

q7="""7.What is the answer to this expression, 22 % 3 is?
a) 7
b) 1
c) 0
d) 5"""

q8="""8.What is the output of this expression, 3*1**3?
a) 27
b) 9
c) 3
d) 1"""

q9="""9. Which one of the following has the highest precedence in the expression?
a) Exponential
b) Addition
c) Multiplication
d) Parentheses"""

q10="""10.Suppose d = {“john”:40, “peter”:45}. To obtain the number of entries in dictionary which command do we use?
a) d.size()
b) len(d)
c) size(d)
d) d.len() """


questions ={q1:'b',q2:'c',q3:'b',q4:'d',q5:'b',q6:'b',q7:'b',q8:'c',q9:'d',q10:'b'}
```

```python
name=input('Hello! Please enter your name:')
score=0
for key in questions:
    print(key)
    choice=input('Do you want to skip it? (yes/no):')
    if choice=='yes':
        print('')
        continue
    response=input("Enter you answer(a/b/c/d):")
    if response==questions[key]:
            print('\033[1;3mCorrect!\033[0m')
            score=score+1
    else:
        print('\033[1;3mIncorrect!\033[0m')
        print('The correct answer is ',questions.get(key))
    print('')

percentage=score/len(questions)*100       
print("Hi",name,'!! Your final score is:',percentage,'%')

if percentage>=65:
    print('\033[1;36mYou have passed the quiz!\033[0m' )
else:
    print('\033[1;36mYou are failed in the quiz!You need to study some basic concepts of python\033[0m')
                     
                  
```
